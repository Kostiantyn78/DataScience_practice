### Analyze the file 2017_jun_final.csv. The file contains the results of the June 2017 developer survey.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the 2017_jun_final.csv file using the "read_csv" method
[LINK](https://docs.google.com/spreadsheets/d/1bclX5PbqE8zR1-wOSM3ZTFa53sIC7RXqwty7Abmw9yQ/edit#gid=1724592523)

```python
url = 'https://docs.google.com/spreadsheets/d/1bclX5PbqE8zR1-wOSM3ZTFa53sIC7RXqwty7Abmw9yQ/export?format=csv&id=1bclX5PbqE8zR1-wOSM3ZTFa53sIC7RXqwty7Abmw9yQ&gid=1724592523'
df = pd.read_csv(url)
print(df)
```

#### Read the resulting table using the "head" method

```python
print(df.head())
```

#### Determine the size of a table using the "shape" method

```python
print(df.shape)
```

#### Determine the types of all columns with "dataframe.dtypes"

```python
print(df.dtypes)
```

#### Calculate the proportion of blanks in each column (use the "isnull" and "sum" methods)

```python
print(df.isnull().sum())
print(df.isnull().sum() / df.shape[0])
```

#### Delete all columns with blanks except for the "Мова програмування" column

```python
# Select columns with missing values
columns_with_mis_val = df.columns[df.isnull().any()].tolist()
print("Columns with missing values", columns_with_mis_val, sep="\n")
# Delete column "Мова програмування" from the list of columns with missing values
columns_to_del = [col for col in columns_with_mis_val if col != "Мова програмування"]
print("Columns to delete", columns_to_del, sep="\n")
# Delete columns
df.drop(columns=columns_to_del, inplace=True)
# Print the modified DataFrame
print(df.head())
```

#### Again, calculate the proportion of spaces in each column and make sure that only the "Мова програмування" column remains.

```python
print(df.isnull().sum())
```

#### Delete all rows in the source table using the "dropna" method

```python
df.dropna(axis=0, inplace=True)
```

#### Define a new table size using the "shape" method

```python
print(df.shape)
```

#### Create a new table python_data, which will contain only rows with specialists who have specified the Python programming language

```python
python_data = df[df["Мова програмування"] == "Python"]
print(python_data.head())
```

#### Determine the size of the python_data table using the "shape" method

```python
print(python_data.shape)
```

```python
salary = python_data[['exp', 'salary']]
plt.figure(figsize=(10, 5))
sns.lineplot(x='exp', y='salary', data=salary)
plt.title('Залежність зарплати Python - спеціаліста від стажу', color='blue')
plt.xlabel('Стаж (роки)', color='blue')
plt.ylabel('Зарплата за місяць', color='blue')
# plt.grid()
plt.show()
```

#### Using the "groupby" method, group by the "Посада" column

```python
groupby_job_title = python_data.groupby("Посада", as_index=False)
print(groupby_job_title.size())
```

#### Create a new DataFrame, where for the grouped data by the "Посада" column, perform data aggregation using the "agg" method and find the minimum and maximum values in the "Зарплата.в.місяць" column

```python
# aggregated_by_salary = groupby_job_title.agg({'Зарплата на місяць': ['min', 'max']})
aggregated_by_salary = groupby_job_title.agg(min_salary=pd.NamedAgg(column="Зарплата на місяць", aggfunc="min"),
                                           max_salary=pd.NamedAgg(column="Зарплата на місяць", aggfunc="max"))
print(aggregated_by_salary.to_string(index=False))
```

#### Create a "fill_avg_salary" function that returns the average salary per month. Use it for the "apply" method and create a new column "avg"

```python
def fill_avg_salary(df_data):
    # return df_data['Зарплата на місяць'].mean()
    return (df_data['min_salary'] + df_data['max_salary']) / 2
    # return df_data.mean()

aggregated_by_salary['avg_salary'] = aggregated_by_salary.apply(fill_avg_salary, axis=1)

print(aggregated_by_salary)
```

```python
plt.figure(figsize=(6, 4))
plt.ylabel('Середня зарплата, USD')
sns.barplot(x='Посада', y='avg_salary', data=aggregated_by_salary, palette='pastel', hue='Посада', dodge=False)
plt.title('Середній рівень зарплати за посадою')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#### Create descriptive statistics using the "describe" method for a new column.

```python
print(aggregated_by_salary['avg_salary'].describe())
```

#### Save the resulting table to a CSV file

```python
aggregated_by_salary.to_csv("aggregated_data.csv")
```
